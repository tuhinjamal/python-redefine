List Comprehension

```python
without_comprehension = []
for i in range(10):
    if(i % 2 == 0):
        without_comprehension.append(i)
print(without_comprehension)
```

```python
Comprehension = [x ** x for x in range(10)]
print(Comprehension)
```

```python
Comprehension = [x for x in range(10) if x%2 !=0]
print(Comprehension)

comp = [x for x in Comprehension if x != 9 ]
print(comp)
```
